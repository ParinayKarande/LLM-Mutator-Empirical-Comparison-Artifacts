```python
import pandas as pd

all_summaries = []

input_file = "/Users/parinaykarande/Desktop/Capstone/Spring_25/Work/PIT_Results_Transformed/PIT_Results_Collections_Transformed.xlsx"
df = pd.read_excel(input_file)

df['Lines Covered (Line Coverage)'] = pd.to_numeric(df['Lines Covered (Line Coverage)'], errors='coerce')
df['Total Lines (Line Coverage)'] = pd.to_numeric(df['Total Lines (Line Coverage)'], errors='coerce')
df['Lines Covered (Mutation)'] = pd.to_numeric(df['Lines Covered (Mutation)'], errors='coerce')
df['Total Lines (Mutation)'] = pd.to_numeric(df['Total Lines (Mutation)'], errors='coerce')


total_covered_line = df['Lines Covered (Line Coverage)'].sum()
total_lines_line = df['Total Lines (Line Coverage)'].sum()
total_covered_mutation = df['Lines Covered (Mutation)'].sum()
total_lines_mutation = df['Total Lines (Mutation)'].sum()


overall_line_coverage = (total_covered_line / total_lines_line) * 100 if total_lines_line else 0
overall_mutation_coverage = (total_covered_mutation / total_lines_mutation) * 100 if total_lines_mutation else 0


# Print totals
print(f"Total Covered (Line Coverage): {total_covered_line}")
print(f"Total Lines (Line Coverage): {total_lines_line}")
print(f"Total Covered (Mutation): {total_covered_mutation}")
print(f"Total Lines (Mutation): {total_lines_mutation}")

print(f"Total Line Coverage: {round(overall_line_coverage)}%")
print(f"Total Mutation Coverage: {round(overall_mutation_coverage)}%")


summary_df = pd.DataFrame([{
    'Project': 'Collections',
    'Total Covered (Line Coverage)': total_covered_line,
    'Total Lines (Line Coverage)': total_lines_line,
    'Total Line Coverage (%)': round(overall_line_coverage,2),
    'Total Covered (Mutation)': total_covered_mutation,
    'Total Lines (Mutation)': total_lines_mutation,
    'Total Mutation Coverage (%)': round(overall_mutation_coverage,2)
}])
all_summaries.append(summary_df.iloc[0].to_dict())
summary_df
```
