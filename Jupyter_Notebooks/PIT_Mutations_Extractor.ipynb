```python
import os
from bs4 import BeautifulSoup
import pandas as pd
import re

def extract_mutations_from_html(file_path):
    mutations = []
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            soup = BeautifulSoup(f, 'html.parser')
            p_tags = soup.find_all('p', class_=['KILLED', 'SURVIVED', 'NO_COVERAGE'])
            if p_tags:
                print(f"[INFO] Found {len(p_tags)} mutation blocks in {os.path.basename(file_path)}")
            for p in p_tags:
                raw_html = str(p)
                # Match content after the last </span></span> up to &rarr;
                match = re.search(r'</span></span>(.*?)→', raw_html)
                if match:
                    mutation = match.group(1).strip()  # Get the matched mutation text
                    mutations.append(mutation)
    except Exception as e:
        print(f"[ERROR] Failed to process {file_path}: {e}")
    return mutations

def collect_all_mutations(root_folder):
    data = []
    print(f"[START] Scanning directory: {root_folder}")
    html_count = 0

    for dirpath, dirnames, filenames in os.walk(root_folder):
        for file in filenames:
            if file.endswith('.html'):
                html_count += 1
                file_path = os.path.join(dirpath, file)
                print(f"[PROCESSING] {file}")
                mutations = extract_mutations_from_html(file_path)
                for mutation in mutations:
                    data.append({'file name': file, 'mutation': mutation})
    
    print(f"\n[COMPLETE] Processed {html_count} HTML files.")
    return pd.DataFrame(data)

# Set folder path
folder_path = '/Users/parinaykarande/Desktop/Capstone/Spring_25/Pitest_Results/commonc-csv-1.13.0'

# Run the extraction
df = collect_all_mutations(folder_path)

# Save to Excel
output_path = '/Users/parinaykarande/Desktop/Capstone/Spring_25/Work/PIT_Mutations/mutations_csv.xlsx'
print(f"\n[SAVING] Writing results to {output_path}")
df.to_excel(output_path, index=False)

print("\n[PREVIEW]")
print(df.head())
```
