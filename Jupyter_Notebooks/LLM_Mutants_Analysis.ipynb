```python
import pandas as pd
from collections import Counter

summary_df = pd.DataFrame()

file_path = '/Users/parinaykarande/Desktop/Capstone/Spring_25/Work/LLM_MutationOperators/MutationOperators_Collections.csv'
df = pd.read_csv(file_path)

mutant_categories = {
    'Conditionals_Boundary': ['<', '<=', '>', '>=', 'Boundary', 'Conditional', 'Bound'],
    'Invert_Negatives': ['Negative','-','!'],
    'Math': ['ReturnValsMutator'],
    'Negate_Conditionals': ['negated conditional','Negate','!=','==','<', '<=', '>', '>='],
    'Return_Values': ['returned', 'return value with ""'],
    'Void_Method_Calls': ['void'],
    'Empty_Returns': ['empty'],
    'False_Returns': ['return with false','false'],
    'True_Returns': ['return with true','true'],
    'Null_Returns': ['null'],
    'Primitive_Returns': ['0','zero']
}


def assign_category(mutation_comment):
    for category, keywords in mutant_categories.items():
        for keyword in keywords:
            if pd.notna(mutation_comment) and keyword.lower() in mutation_comment.lower():
                return category
    return 'Unknown'


explored_comments = df['Mutation Comments'].dropna().apply(lambda x: x.split('\n')).explode().reset_index(drop=True)

explored_comments = explored_comments[explored_comments.str.strip() != '']  # remove blank lines
categorized = explored_comments.apply(assign_category)

category_counts = categorized.value_counts()

category_dict = category_counts.to_dict()
category_dict['Project'] = "Collections"

summary_df = pd.concat([summary_df, pd.DataFrame([category_dict])], ignore_index=True)

print(summary_df)
```
