```python
import pandas as pd

all_summaries = []

input_file = '/Users/parinaykarande/Desktop/Capstone/Spring_25/Work/LLM_Results/LLM_Results_Collections.xlsx'
df = pd.read_excel(input_file)

df['TOTAL_LINES_PRE'] = df['LINE_MISSED_PRE_MUTATION'] + df['LINE_COVERED_PRE_MUTATION']

df['TOTAL_LINES_POST'] = df['LINE_MISSED_POST_MUTATION'] + df['LINE_COVERED_POST_MUTATION']

df['Coverage_Pre_Mutation'] = (df['LINE_COVERED_PRE_MUTATION'] / df['TOTAL_LINES_PRE']) * 100
df['Coverage_Post_Mutation'] = (df['LINE_COVERED_POST_MUTATION'] / df['TOTAL_LINES_POST']) * 100


total_lines1 = df['TOTAL_LINES_PRE'].sum()
total_lines2 = df['TOTAL_LINES_POST'].sum()
total_covered_pre = df['LINE_COVERED_PRE_MUTATION'].sum()
total_covered_post = df['LINE_COVERED_POST_MUTATION'].sum()

# Calculate overall percentages
overall_coverage_pre = (total_covered_pre / total_lines1) * 100
overall_coverage_post = (total_covered_post / total_lines2) * 100

# Print
print(f"Total Lines 1: {total_lines1}")
print(f"Total Lines 2: {total_lines2}")
print(f"Total Covered Pre-Mutation: {total_covered_pre}")
print(f"Total Covered Post-Mutation: {total_covered_post}")


print(f"Overall Coverage Pre-Mutation: {overall_coverage_pre:.2f}%")
print(f"Overall Coverage Post-Mutation: {overall_coverage_post:.2f}%")



#Pseudo Mutation Coverage
if overall_coverage_pre > overall_coverage_post:
    lam_score = ((overall_coverage_pre - overall_coverage_post) / overall_coverage_pre) * 100
else:
    lam_score = 0

# Print the result
print(f" LAM Score: {lam_score:.2f}%")
print(f" The test suite successfully killed {lam_score:.2f}% of mutants introduced by LLM mutations.")

summary_df = pd.DataFrame([{
    'Project': 'Collections',
    'Total Lines Pre-Mutation': total_lines1,
    'Total Lines Post-Mutation': total_lines2,
    'Covered Pre-Mutation': total_covered_pre,
    'Covered Post-Mutation': total_covered_post,
    'Coverage Pre-Mutation (%)': round(overall_coverage_pre, 2),
    'Coverage Post-Mutation (%)': round(overall_coverage_post, 2),
    'LAM Score (%)': round(lam_score, 2)
}])
summary_df
all_summaries.append(summary_df.iloc[0].to_dict())
```
