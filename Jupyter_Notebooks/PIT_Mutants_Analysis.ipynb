```python
import pandas as pd
from collections import Counter

summary_df = pd.DataFrame()

file_path = '/Users/parinaykarande/Desktop/Capstone/Spring_25/Work/PIT_Mutations/mutations_collections.xlsx'
df = pd.read_excel(file_path)

mutant_categories = {
    'Conditionals_Boundary': ['<', '<=', '>', '>=', 'Boundary', 'Conditional', 'Bound'],
    'Invert_Negatives': ['Negative','-','!'],
    'Math': ['ReturnValsMutator'],
    'Negate_Conditionals': ['negated conditional','!=','==','<', '<=', '>', '>='],
    'Return_Values': ['returned', 'return value with ""'],
    'Void_Method_Calls': ['void'],
    'Empty_Returns': ['empty'],
    'False_Returns': ['return with false','false'],
    'True_Returns': ['return with true','true'],
    'Null_Returns': ['null'],
    'Primitive_Returns': ['0','zero']
}


def assign_category(mutation_comment):
    for category, keywords in mutant_categories.items():
        for keyword in keywords:
            if pd.notna(mutation_comment) and keyword.lower() in mutation_comment.lower():
                return category
    return 'Unknown'

df['Category'] = df['mutation'].apply(assign_category)

category_counts = df['Category'].value_counts()

category_dict = category_counts.to_dict()
category_dict['Project'] = "Collections"

summary_df = pd.concat([summary_df, pd.DataFrame([category_dict])], ignore_index=True)

print(summary_df)
```
